```python
import numpy as np
import matplotlib.pyplot as plt

def maxpool_rgb(image, kernel_size=2, stride=2):
    """
    Fhrt MaxPooling auf einem RGB-Bild durch.
    - image: Eingangsbild (H, W, C)
    - kernel_size: Fenstergre (z.B. 2 oder 3)
    - stride: Schrittweite (Standard = kernel_size -> nicht berlappend)
    """
    h, w, c = image.shape
    out_h = (h - kernel_size) // stride + 1
    out_w = (w - kernel_size) // stride + 1

    # Output-Bild
    pooled = np.zeros((out_h, out_w, c), dtype=image.dtype)

    for i in range(out_h):
        for j in range(out_w):
            # Fenster extrahieren
            window = image[
                i*stride : i*stride+kernel_size,
                j*stride : j*stride+kernel_size,
                :
            ]
            # Kanalweise Maximum
            pooled[i, j, :] = window.max(axis=(0,1))
    
    return pooled

# Beispiel: Bild laden (wie bei dir)
WIDTH = 640
HEIGHT = 480
with open("input_image.dat", "rb") as f:
    raw_data = np.frombuffer(f.read(), dtype=np.uint8)

image = raw_data.reshape((HEIGHT, WIDTH, 3))

# MaxPooling ausfhren (z.B. 3x3, stride 3)
pool_result = maxpool_rgb(image, kernel_size=2, stride=2)

# Plot
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(image)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("MaxPooling 3x3 (stride=3)")
plt.imshow(pool_result)
plt.axis("off")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def read_dat_image(dat_filename, width, height):
    """
    Liest eine .dat Datei mit 24-bit RGB (3 Bytes pro Pixel) ein
    und gibt ein NumPy Array der Form (height, width, 3) zurck.
    """
    expected_size = width * height * 3
    with open(dat_filename, "rb") as f:
        raw_data = f.read()
    if len(raw_data) != expected_size:
        raise ValueError(f"Dateigre passt nicht zum Bildformat: erwartet {expected_size} Bytes, Datei hat {len(raw_data)} Bytes")
    # In NumPy Array umwandeln und reshape
    img = np.frombuffer(raw_data, dtype=np.uint8)
    img = img.reshape((height, width, 3))
    plt.imshow(img)
    return img

def output_size(input_size, kernel_size, stride, padding=0):
    return ((input_size - kernel_size + 2*padding) // stride) + 1

    
# Beispielnutzung:
max_pooling_result_vhdl = read_dat_image("output_image_Max_Pooling.dat", output_size(640,2,2,0), output_size(480,2,2,0))
# img_b = read_dat_image("image2.dat", 640, 480)
# compare_images(img_a, img_b)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

def compare_images(img1, img2, img3, diff_color='red'):


    assert img1.shape == img2.shape, "Die Bilder mssen die gleiche Form haben!"

    # Differenz berechnen (absolut)
    diff = np.abs(img1.astype(np.int16) - img2.astype(np.int16))

    # Max und Mittelwerte pro Kanal
    max_diff = diff.max(axis=(0,1))
    mean_diff = diff.mean(axis=(0,1))

    print("Maximale Differenz pro Kanal (R,G,B):", max_diff)
    print("Mittlere Differenz pro Kanal (R,G,B):", mean_diff)

    # Erstelle einen einfarbigen Colormap Verlauf (wei -> Farbe)
    if diff_color == 'gray':
        cmap = plt.cm.gray
    else:
        # custom cmap von wei zu diff_color
        cmap = LinearSegmentedColormap.from_list('custom_cmap', ['white', diff_color])

    fig, axs = plt.subplots(2, 3, figsize=(24, 10))
    axs = axs.flatten()

    # Bild 1
    axs[0].set_title("Original Bild")
    axs[0].imshow(img3)
    axs[0].axis('off')

    # Bild 2
    axs[1].set_title("Covolution Keras")
    axs[1].imshow(img1)
    axs[1].axis('off')

    # Bild 3
    axs[2].set_title("Convolution VHDL")
    axs[2].imshow(img2)
    axs[2].axis('off')


    # Einzelne Kanle der Differenz, normiert, in gleicher Farbe
    for i, color_name in enumerate(['Rot', 'Grn', 'Blau']):
        channel_img = diff[:,:,i]
        max_channel = channel_img.max()
        scaled = channel_img / max_channel if max_channel > 0 else channel_img*0

        # Bild als 2D, aber mit Colormap (1-Kanal Bild)
        im = axs[3+i].imshow(scaled, cmap=cmap)
        axs[3+i].set_title(f"Absolute Differenz {color_name}")
        axs[3+i].axis('off')

        cbar = plt.colorbar(im, ax=axs[3+i], fraction=0.046, pad=0.04)
        cbar.set_label("Rel. Differenz (normiert)")

    # Leeres 8. Feld (fr Layout)
    #axs[7].axis('off')

    plt.tight_layout()
    # Bild speichern
    plt.savefig("vergleich_plot.png", dpi=300, bbox_inches='tight')  # png, hohe Auflsung, sauber zugeschnitten

    plt.show()

# Beispielaufruf:
# compare_images(img1, img2, diff_color='blue')
```
